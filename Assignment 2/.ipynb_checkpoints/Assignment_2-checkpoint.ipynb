## Construct Gaussian Generative Model classifier for each of the 3 classes.
#### Test each classifier on all images in X_Test. 
#### Construct Confusion Matrix.

## Comment on the results ##

### 
~ According to the outcome, Gaussian generative model achieved higher accuracy than Naive Bayes classifier because it doesn't assume independency between pixel values while the Naive Bayes does. For sure, for a shape to be defined we have to look at a group of pixels not individual pixels, this means that a shape is defined by the relation between its pixels not by the values of its individual pixels only.

~ As for the difference in accuaracies between different image sizes, A possible explanation for the 32x32 being the least accuracy is that we have exceeded the optimal number of features, that it's now overfitting, meaning that the model started learning exceptions that are specific to the training data and do not generalize well when new data is encountered.

~ A possible explanation for the decrease in accuracy of image sizes 16x16 to 4x4 is that the images become less separable when we decrease the number of data features.
